```python
import pandas as pd
```

```python
data = pd.read_csv(r'/content/titanic.csv')
data.head()
```

```python
df= data.drop(columns= ['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'])
df.head()
```

```python
IV = df.drop('Survived', axis = 'columns')
TV = df['Survived']
IV.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Le_Class = LabelEncoder()
Le_Sex = LabelEncoder()
Le_Age = LabelEncoder()
Le_Fare = LabelEncoder()
```

```python
IV['Class'] = Le_Class.fit_transform(IV['Pclass'])
IV['Sex_p'] = Le_Sex.fit_transform(IV['Sex'])
IV['Age_p'] = Le_Age.fit_transform(IV['Age'])
IV['Fare_P'] = Le_Fare.fit_transform(IV['Fare'])
```

```python
IV.head()
```

```python
IV2 = IV.drop(['Pclass','Sex','Age','Fare'], axis = 'columns')
```

```python
IV2.head()
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(IV2,TV)
```

```python
model.score(IV2, TV)
```
